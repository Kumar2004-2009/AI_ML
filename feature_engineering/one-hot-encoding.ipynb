Nominal categorical data to numerical data using One Hot Encoding

Dummy Variable trap - remove the first column
why?

to avoid multicollinearity (the input columns must not have the dependency with each other)



Use the frequent categories and rest of them make it as the column others

```
import numpy as np
import pandas as pd
```

```
df=pd.read_csv("cars.csv")
```

```
df.head()
```

```
df['brand'].value_counts()
```

1. One Hot encoding using Pandas - it does not remeber the which column plced at which position

```
pd.get_dummies(df,columns=['fuel','owner'])
```

K-1 One Hot Encoding

```
pd.get_dummies(df,columns=['fuel','owner'],drop_first=True)
#
```

2. Using Sklearn

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=2)
```

```
from sklearn.preprocessing import OneHotEncoder
```

```
ohe = OneHotEncoder(drop='first',dtype=np.int32)
```

```
X_train_new= ohe.fit_transform(X_train[['fuel','owner']]).toarray()
X_test_new = ohe.transform(X_test[['fuel','owner']]).toarray()
```

```
np.hstack((X_train[['brand','km_driven']].values,X_train_new)).shape
```

3. One Hot encoding with Top categories - use it when some categories have more frequency than others

```
counts=df['brand'].value_counts()
```

```
df['brand'].nunique()
threshold=100
```

```
repl=counts[counts <= threshold].index
```

```
pd.get_dummies(df['brand'].replace(repl,'uncommon')).sample(5)
```
