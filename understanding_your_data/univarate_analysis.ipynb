```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("train.csv")
```

```python
df.head()
```

```python
df.info()
```

1. Categorical Data

a. Countplot

```python
sns.countplot(x=df['Survived'])
```

b. Pie Chart

```python
df['Survived'].value_counts().plot(kind='pie',autopct='%.2f')
```

2. Numerical Data

a. Histrogram

```python
plt.hist(df['Age'],bins=15)
```

b. Distplot

```python
sns.distplot(x=df['Age'])
```

c. Boxplot

```python
# sns.boxplot(df['Fare'])
sns.boxplot(df['Age'])
```
